```python
import subprocess
import time
import os

os.environ['CUDA_VISIBLE_DEVICES'] = '0,1'
os.environ['OLLAMA_HOST'] = '0.0.0.0:11434'
os.environ['LD_LIBRARY_PATH'] = '/usr/local/cuda/lib64:' + os.environ.get('LD_LIBRARY_PATH', '')

# Дополнительные параметры Ollama
os.environ['OLLAMA_NUM_PARALLEL'] = '1'       # Один запрос за раз (меньше OOM)
os.environ['OLLAMA_MAX_LOADED_MODELS'] = '1'  # Держать только одну модель в памяти
os.environ['OLLAMA_KEEP_ALIVE'] = '-1'        # Не выгружать модель между запросами!

# Ограничь контекстное окно при запуске модели
# os.environ['OLLAMA_NUM_CTX'] = '4096'  # вместо дефолтного 128k у gemma4

log_file = open('/tmp/ollama.log', 'w')

process = subprocess.Popen(
    ['ollama', 'serve'],
    # stdout=subprocess.PIPE,
    # stderr=subprocess.PIPE,
    stdout=log_file,
    stderr=log_file,
)

time.sleep(8)
print("✅ Ollama запущена!")

# Проверяем, что сервер видит GPU
result = subprocess.run(['ollama', 'list'], capture_output=True, text=True)
print(result.stdout)
```

```python
from openai import OpenAI

# Подключаемся к локальной Ollama
# api_key может быть любым (например, "ollama"),
# так как локальная модель не требует авторизации,
# но параметр обязателен для клиента OpenAI.
client = OpenAI(base_url="http://localhost:11434/v1/", api_key="ollama")

# Имя модели, которую вы скачали (можно посмотреть командой ollama list)
model_name = "gemma4:latest"

print(f"Отправка запроса к модели {model_name}...\n")

try:
    # Делаем запрос к модели (Chat Completions)
    response = client.chat.completions.create(
        model=model_name,
        messages=[
            {
                "role": "user",
                "content": "Почему небо голубое? Ответь в двух предложениях.",
            },
        ],
        temperature=0.7,
        max_tokens=10000,
    )

    # Выводим ответ
    # print("response:", response)
    answer = response.choices[0].message.content
    print("Ответ модели:")
    print(answer)

except Exception as e:
    print(f"Произошла ошибка: {e}")
```
